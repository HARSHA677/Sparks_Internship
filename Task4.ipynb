# Findings

```python
print("Country with the most attack: ",data['Country'].value_counts().idxmax())
print('City with most attack: ',data['city'].value_counts().index[1])
print('Region with most attack:',data['Region'].value_counts().idxmax())
print('state with most attack:',data['state'].value_counts().idxmax())
print('Year with most attack: ',data['Year'].value_counts().idxmax())
print('Month with most attack: ',data['Month'].value_counts().idxmax())
print('Group with most attack :',data['Group'].value_counts().index[1])
print('Most attack type: ',data['AttackType'].value_counts().idxmax())
```

# Checking whether total no of Terror attacks increased or decreased 




```python
plt.figure(figsize=(10,6))
fig = sns.distplot(data["Year"].values, color = 'y')
fig.set_xlabel(".Year",size=15)
fig.set_ylabel("Terror attack total",size=15)
plt.title('Number of Terror attacks each year',size = 20)
plt.show()
```

# Yearly Plot

```python
data['Year'].value_counts(dropna=False).sort_index()
```

```python
plt.subplots(figsize=(18,8))
sns.countplot('Year',data = data)
sns.color_palette("husl",8)
plt.xticks(rotation=45)
plt.title('Number of Terrorist Activites Each Year',fontsize=20)
plt.show()
```

# By Country

```python
plt.subplots(figsize=(18,6))
country_attacks = data.Country.value_counts()[:15].reset_index()
country_attacks.columns = ['Country','Total Attacks']
sns.barplot(x=country_attacks.Country, y=country_attacks['Total Attacks'], palette='husl',edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=30)
plt.title('Number of Total Attacks in Each Country',fontsize=20)
plt.show()
```

# By region

```python
plt.subplots(figsize=(18,6))
Region_attacks = data.Region.value_counts()[:15].reset_index()
Region_attacks.columns = ['Region','Total Attacks']
sns.barplot(x=Region_attacks.Region, y=Region_attacks['Total Attacks'], palette='husl',edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=30)
plt.title('Number of Total Attacks in Each Region',fontsize=20)
plt.show()
```

# Total no of Casualities in each country

```python
data['Wounded']=data['Wounded'].fillna(0).astype(int)
data['Killed'] = data['Killed'].fillna(0).astype(int)
data['Casualities'] = data['Killed'] + data['Wounded']
```

```python
plt.subplots(figsize=(15,6))
count_cas = data.groupby('Country').Casualities.sum().to_frame().reset_index().sort_values('Casualities', ascending=False)[:15]
sns.barplot(x=count_cas.Country, y=count_cas.Casualities, palette= 'husl',edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=30)
plt.title('Number of Total Casualities in Each Country',fontsize=20)
plt.show()
```

# Group with highest terror attack

```python
sns.barplot(data['Group'].value_counts()[1:15].values,data['Group'].value_counts()[1:15].index,palette=('husl'))
plt.xticks(rotation=90)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.title('Terrorist Groups with Highest Terror Attacks',fontsize=20)
plt.show()
```

# Attack Type

```python
label1=data['AttackType'].value_counts()[:5].index
size=data['AttackType'].value_counts()[:5].values
explode = (0.1, 0, 0, 0,0) 
fig1, ax1 = plt.subplots()
ax1.pie(size, explode=explode, labels=label1, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Armed assault and bombing/explosion are seen to be the cause of 77% of the deaths in these attacks. This rate is why these attacks are used so many times in terrorist actions. This is how dangerous weapons and explosives are to the world.

# Trend of attack in each region


```python
plt.rcParams['figure.figsize']=18,7
terror_area=pd.crosstab(data.Year,data.Region)
terror_area.plot()
plt.title('Trend of terror attacks in each region',fontsize=20)
plt.show()
```

# Top 10 countries affected by Terrorism


```python
data.groupby('Country')['Killed'].sum().sort_values(ascending=False)[:10]
```

```python
plt.subplots(figsize=(15,6))
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values,palette='husl')
plt.title('Top Countries Affected',fontsize=16)
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

# No of attacks and casualities by city

```python
city_attacks = data.city.value_counts().to_frame().reset_index()
city_attacks.columns = ['city', 'Total Attacks']
city_cas = data.groupby('city').Casualities.sum().to_frame().reset_index()
city_cas.columns = ['city', 'Casualities']
# city_cas.drop('Unknown', axis=0, inplace=True)
city_tot = pd.merge(city_attacks, city_cas, how='left', on='city').sort_values('Total Attacks', ascending=False)[1:21]
```

```python
sns.set_palette('husl')
city_tot.plot.bar(x='city', width=0.8)
plt.xticks(rotation=90)
plt.title('Number Of Total Attacks and Casualities by City',fontsize=20)
fig = plt.gcf()
fig.set_size_inches(16,9)
plt.show()
```
